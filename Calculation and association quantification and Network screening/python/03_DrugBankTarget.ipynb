```python
import json
import xmltodict
import re

# 读取 JSON 数据
with open('drugBank_5_1_11.json', 'r', encoding="utf-8") as f:
    data = json.load(f)

drugs = data['drugbank']['drug']
file = open('items.txt', 'w', encoding="utf-8")
```

```python
for drug in drugs:
    try:
        drug_id = drug['drugbank-id'][0]['#text']
    except:
        drug_id = drug['drugbank-id']['#text']
    drug_name = drug['name']
    try:
        targets = drug['targets']['target']
    except:
        continue

    drug_type = drug['@type']

    if isinstance(targets, dict):
        try:
            uniprot_id = targets['polypeptide']['@id']
        except:
            continue
        target_id = targets['id']
        target_name = targets['name']
        # target_name  = re.sub(r',',"-",target_name2)



        try:
            target_actioncar = targets['actions']['action']
            target_action = ''.join(target_actioncar)
        except:
            continue
        try:
            target_geneName = targets['polypeptide']['gene-name']
        except:
            continue
        

        if target_action is None:
            target_action = "WU"


        if target_geneName is None:
            target_geneName = "WU"


        target_aminoAcidSequenceFormat = targets['polypeptide']['amino-acid-sequence']['@format']

        try:
            target_aminoAcidSequenceText2 = targets['polypeptide']['amino-acid-sequence']['#text']
            target_aminoAcidSequenceText  = re.sub(r'\n',"/n",target_aminoAcidSequenceText2)
        except:
            continue

        if target_aminoAcidSequenceText is None:
            target_aminoAcidSequenceText = "WU"

        try:
            target_geneSequenceText2 = targets['polypeptide']['gene-sequence']['#text']
            target_geneSequenceText  = re.sub(r'\n',"/n",target_geneSequenceText2)
        except:
            continue

        if target_geneSequenceText is None:
            target_geneSequenceText = "WU"

        target_geneS2quenceFormat = targets['polypeptide']['gene-sequence']['@format']
        
        item = '\t'.join([drug_id, drug_name, drug_type, uniprot_id, target_name, target_action, target_geneName, target_aminoAcidSequenceFormat, target_aminoAcidSequenceText, target_geneS2quenceFormat, target_geneSequenceText])
        file.writelines(item + '\n')

    if isinstance(targets, list):
        for target in targets:
            try:
                uniprot_id = target['polypeptide']['@id']
            except:
                continue
            target_id = target['id']
            target_name = target['name']
            # target_name  = re.sub(r',',"-",target_name2)
            try:
                target_actioncar = target['actions']['action']
                target_action = ''.join(target_actioncar)
            except:
                continue
            try:
                target_geneName = target['polypeptide']['gene-name']
            except:
                continue
            

            if target_action is None:
                target_action = "WU"


            if target_geneName is None:
                target_geneName = "WU"

            try:
                target_aminoAcidSequenceText2 = target['polypeptide']['amino-acid-sequence']['#text']
                target_aminoAcidSequenceText  = re.sub(r'\n',"/n",target_aminoAcidSequenceText2)
            except:
                continue
            if target_aminoAcidSequenceText is None:
                target_aminoAcidSequenceText = "WU"
            try:
                target_geneSequenceText2 = target['polypeptide']['gene-sequence']['#text']
                target_geneSequenceText  = re.sub(r'\n',"/n",target_geneSequenceText2)
            except:
                continue
            if target_geneSequenceText is None:
                target_geneSequenceText = "WU"



            target_aminoAcidSequenceFormat = target['polypeptide']['amino-acid-sequence']['@format']
            target_geneS2quenceFormat = target['polypeptide']['gene-sequence']['@format']
            item = '\t'.join([drug_id, drug_name, drug_type, uniprot_id, target_name, target_action, target_geneName, target_aminoAcidSequenceFormat, target_aminoAcidSequenceText, target_geneS2quenceFormat, target_geneSequenceText])
            file.writelines(item + '\n')

file.close()
```
